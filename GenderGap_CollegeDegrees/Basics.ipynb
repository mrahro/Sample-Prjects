```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
fig = plt.figure(figsize=(16, 20))

for sp in range(0,18):
    if sp!=16:
        ax = fig.add_subplot(6,3,sp+1)
    if (sp%3)==0:
        integer=int(sp/3)
        ax.plot(women_degrees['Year'], women_degrees[stem_cats[integer]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[integer]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(stem_cats[integer])
        if sp == 0:
            ax.text(2005, 10, 'Men')
            ax.text(2005, 87, 'Women')
    elif (sp%3)==1:
        integer=int(sp/3)
        if sp!=16:
            ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[integer]], c=cb_dark_blue, label='Women', linewidth=3)
            ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[integer]], c=cb_orange, label='Men', linewidth=3)
            ax.set_title(lib_arts_cats[integer])
            if sp == 1:
                ax.text(2007, 20, 'Men')
                ax.text(2005, 75, 'Women')       
    elif (sp%3)==2:
        integer=int(sp/3)
        ax.plot(women_degrees['Year'], women_degrees[other_cats[integer]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[integer]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(other_cats[integer])
        if sp == 2:
            ax.text(2007, 8, 'Men')
            ax.text(2005, 92, 'Women')
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_yticks([0,100])
    ax.set_ylim(0,100)
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    if (sp!=15) and (sp!=13) and (sp!=17):
        ax.tick_params(labelbottom='off')
        
    plt.savefig('gender_degrees.png')
    
plt.show()
```
